```python
import numpy as np
```

```python
universe = """...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#.....
"""

universe = open("input.txt").read()
```

```python
um = np.array([[char for char in line] for line in universe.splitlines()])
print(um)
```

```python
ys, xs = np.where(um == "#")
dbys = set(i for i in range(0, um.shape[0])) - set(ys)
dbxs = set(i for i in range(0, um.shape[1])) - set(xs)
print(dbxs, dbys)
```

```python
from itertools import combinations

d_sum = 0

for a, b in combinations(zip(ys, xs), 2):
    d_y = sum(
        (1000000 if y in dbys else 1 for y in range(min(a[0], b[0]), max(a[0], b[0])))
    )
    d_x = sum(
        (1000000 if x in dbxs else 1 for x in range(min(a[1], b[1]), max(a[1], b[1])))
    )
    d_sum += d_y + d_x
```
